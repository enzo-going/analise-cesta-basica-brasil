```python
import re

def convert_to_float(time_str):
  """Converts a string in 'HHhMMm' format to a float 'HH.MM'."""
  try:
    match = re.match(r'(\d+)h(\d+)m', time_str)
    if match:
      hours = int(match.group(1))
      minutes = int(match.group(2))
      return float(f"{hours}.{minutes:02d}")
    else:
      return 0.0
  except (ValueError, AttributeError, TypeError):
    return 0.0

for col in df.columns:
  if(col == 'mês-ano'): continue
  df[col] = df[col].apply(convert_to_float)

print(df)
```

# Tabela Horas Trabalhadas (Média, Mediana e Desvio Geral)



```python
import pandas as pd
import numpy as num
from google.colab import data_table
data_table.enable_dataframe_formatter

def tempo_para_minutos(tempo_str):
    """Converte tempo em formato XXhXXm para minutos totais"""
    try:
        if pd.isna(tempo_str) or tempo_str == '':
            return None

        tempo_str = str(tempo_str).strip()

        # Extrair horas e minutos do formato "145h04m"
        if 'h' in tempo_str and 'm' in tempo_str:
            partes = tempo_str.replace('m', '').split('h')
            horas = int(partes[0])
            minutos = int(partes[1])
            return horas * 60 + minutos
        else:
            return None
    except:
        return None

def minutos_para_tempo(minutos):
    """Converte minutos totais de volta para formato XXhXXm"""
    if pd.isna(minutos) or minutos is None:
        return None

    minutos = int(minutos)
    h = minutos // 60
    m = minutos % 60
    return f"{h}h{m:02d}m"

# Carregar arquivo
arquivo = '/content/drive/MyDrive/exporta.xls'
df = pd.read_excel(arquivo)
df.columns = df.loc[0]
df.columns.values[0] = 'mes-ano'
df = df[1:-3].replace(to_replace=r'\r', value='', regex=True)

# Exibir dados originais
print("📊 DADOS ORIGINAIS:")
display(df)
# Filtrar dados a partir de 1998
df_filtered = df[df['mes-ano'].str.contains(r'(199[5-9]|20[0-9]{2})', na=False)]
print("DADOS FILTRADOS A PARTIR DE 1995:")
display(df_filtered)
df = df_filtered.copy()

# Vamos pegar todas as colunas a partir da 3ª (índice 2)
colunas_cidades = []
for i, col in enumerate(df.columns):
    # Pular apenas a primeira coluna (mes-ano)
    if i >= 1 and col is not None and str(col).strip() != '':
        # Verificar se não é uma coluna 'Unnamed'
        if not str(col).startswith('Unnamed'):
            colunas_cidades.append(col)

print(f"\n🏙️ Cidades identificadas: {colunas_cidades}")

# Processar cada cidade
resultados = []

for cidade in colunas_cidades:
    # Converter tempos para minutos
    minutos = df[cidade].apply(tempo_para_minutos)
    minutos_validos = minutos.dropna()

    if not minutos_validos.empty:
        media_min = minutos_validos.mean()
        mediana_min = minutos_validos.median()
        std_min = minutos_validos.std()

        resultados.append({
            'Cidade': cidade,
            'Registros': len(minutos_validos),
            'Média': minutos_para_tempo(media_min),
            'Mediana': minutos_para_tempo(mediana_min),
            'Desvio Padrão': minutos_para_tempo(std_min),
            'Mínimo': minutos_para_tempo(minutos_validos.min()),
            'Máximo': minutos_para_tempo(minutos_validos.max())
        })

# Criar DataFrame com resultados
if resultados:
    estatisticas = pd.DataFrame(resultados)

    print("\n📈 ESTATÍSTICAS POR CIDADE:")
    display(estatisticas)

    # Calcular estatística geral
    todos_minutos = []
    for cidade in colunas_cidades:
        minutos = df[cidade].apply(tempo_para_minutos).dropna()
        todos_minutos.extend(minutos.tolist())

    if todos_minutos:
        geral = pd.DataFrame({
            'Métrica': ['Geral (Todas as Cidades)'],
            'Total de Registros': [len(todos_minutos)],
            'Média Geral': [minutos_para_tempo(np.mean(todos_minutos))],
            'Mediana Geral': [minutos_para_tempo(np.median(todos_minutos))],
            'Desvio Padrão Geral': [minutos_para_tempo(np.std(todos_minutos))]
        })

        print("\n🌎 ESTATÍSTICA GERAL:")
        display(geral)

        #Display de média de horas por período
        print("\n📅 MÉDIA DE HORAS POR PERÍODO:")

        # Extrair ano da coluna mes-ano
        df_periodo = df.copy()
        df_periodo['ano'] = df_periodo['mes-ano'].str.extract(r'(\d{4})')

        # Calcular média por ano
        medias_periodo = []

        for ano in sorted(df_periodo['ano'].dropna().unique()):
            df_ano = df_periodo[df_periodo['ano'] == ano]

            # Calcular média de todas as cidades para este ano
            minutos_ano = []
            for cidade in colunas_cidades:
                minutos_cidade = df_ano[cidade].apply(tempo_para_minutos).dropna()
                minutos_ano.extend(minutos_cidade.tolist())

            if minutos_ano:
                media_ano = np.mean(minutos_ano)
                medias_periodo.append({
                    'Ano': int(ano),
                    'Média de Horas': minutos_para_tempo(media_ano),
                    'Registros': len(minutos_ano)
                })

        if medias_periodo:
            df_medias_periodo = pd.DataFrame(medias_periodo)
            display(df_medias_periodo)
    else:
      print("❌ Não foi possível processar os dados")
      print("Estrutura das colunas:", df.columns.tolist())
      print("Primeiras linhas:")
      display(df.head())
```

#Tabela de Preços (Média, Mediana e Desvio Geral)

```python
import pandas as pd
import numpy as np


def limpar_valor_monetario(valor):
    """Converte valores monetários para float, tratando diferentes formatos"""
    try:
        if pd.isna(valor) or valor == '':
            return None

        # Converter para string e limpar
        valor_str = str(valor).strip()

        # Remover símbolos de moeda e espaços
        valor_limpo = valor_str.replace('R$', '').replace('$', '').replace(' ', '')

        # Tratar vírgula como separador decimal (formato brasileiro)
        if ',' in valor_limpo and '.' in valor_limpo:
            # Se tem ambos, vírgula é separador de milhar, ponto é decimal
            valor_limpo = valor_limpo.replace(',', '')
        elif ',' in valor_limpo:
            # Se só tem vírgula, é separador decimal
            valor_limpo = valor_limpo.replace(',', '.')

        return float(valor_limpo)
    except:
        return None

def formatar_moeda(valor):
    """Formata valor como moeda brasileira"""
    if pd.isna(valor) or valor is None:
        return None
    return f"R$ {valor:,.2f}".replace(',', 'X').replace('.', ',').replace('X', '.')

# Carregar arquivo
arquivo = '/content/drive/MyDrive/precocesta.xlsx'  # Ajuste o caminho conforme necessário
df = pd.read_excel(arquivo)

# Configurar cabeçalhos
df.columns = df.iloc[0]  # Usar primeira linha como cabeçalho
df.columns.values[0] = 'mes-ano'  # Renomear primeira coluna
df = df[1:].reset_index(drop=True)  # Remover linha de cabeçalho e resetar índice

# Limpar dados (remover quebras de linha)
df = df.replace(to_replace=r'\r|\n', value='', regex=True)

# Exibir dados originais
print("📊 DADOS ORIGINAIS:")
display(df)

# Filtrar dados a partir de 1995 (modificação principal)
df_filtered = df[df['mes-ano'].str.contains(r'(199[5-9]|20[0-9]{2})', na=False)]
print("DADOS FILTRADOS A PARTIR DE 1995:")
df_filtered

# Usar dados filtrados para análise
df = df_filtered.copy()

# Identificar colunas de cidades (todas exceto a primeira)
colunas_cidades = []
for i, col in enumerate(df.columns):
    if i >= 1 and col is not None and str(col).strip() != '':
        if not str(col).startswith('Unnamed'):
            colunas_cidades.append(col)

print(f"\n🏙️ Cidades identificadas: {colunas_cidades}")

# Processar cada cidade
resultados = []

for cidade in colunas_cidades:
    # Converter valores para float
    valores = df[cidade].apply(limpar_valor_monetario)
    valores_validos = valores.dropna()

    if not valores_validos.empty:
        media = valores_validos.mean()
        mediana = valores_validos.median()
        desvio_padrao = valores_validos.std()
        minimo = valores_validos.min()
        maximo = valores_validos.max()

        resultados.append({
            'Cidade': cidade,
            'Registros': len(valores_validos),
            'Média': formatar_moeda(media),
            'Mediana': formatar_moeda(mediana),
            'Desvio Padrão': formatar_moeda(desvio_padrao),
            'Mínimo': formatar_moeda(minimo),
            'Máximo': formatar_moeda(maximo),
            'Soma Total': formatar_moeda(valores_validos.sum())
        })

# Criar DataFrame com resultados
if resultados:
    estatisticas = pd.DataFrame(resultados)

    print("\n📈 ESTATÍSTICAS POR CIDADE (A PARTIR DE 1995):")
    display(estatisticas)

    # Calcular estatística geral
    todos_valores = []
    for cidade in colunas_cidades:
        valores = df[cidade].apply(limpar_valor_monetario).dropna()
        todos_valores.extend(valores.tolist())

    if todos_valores:
        geral = pd.DataFrame({
            'Métrica': ['Geral (Todas as Cidades)'],
            'Total de Registros': [len(todos_valores)],
            'Média Geral': [formatar_moeda(np.mean(todos_valores))],
            'Mediana Geral': [formatar_moeda(np.median(todos_valores))],
            'Desvio Padrão Geral': [formatar_moeda(np.std(todos_valores))],
            'Valor Total': [formatar_moeda(np.sum(todos_valores))],
            'Valor Mínimo': [formatar_moeda(np.min(todos_valores))],
            'Valor Máximo': [formatar_moeda(np.max(todos_valores))]
        })

        print("\n🌎 ESTATÍSTICA GERAL (A PARTIR DE 1995):")
        display(geral)

        # Análise adicional por período
        print("\n📅 ANÁLISE TEMPORAL (A PARTIR DE 1995):")
        df_temporal = df.copy()

        # Converter valores monetários
        for cidade in colunas_cidades:
            df_temporal[cidade] = df_temporal[cidade].apply(limpar_valor_monetario)

        # Calcular média por período
        df_temporal['Média_Período'] = df_temporal[colunas_cidades].mean(axis=1)
        df_temporal_display = df_temporal[['mes-ano', 'Média_Período']].copy()
        df_temporal_display['Média_Período'] = df_temporal_display['Média_Período'].apply(formatar_moeda)

        print("Média por período:")
        display(df_temporal_display)
        df_temporal.to_html()

    else:
        print("❌ Não foi possível processar os dados")
        print("Estrutura das colunas:", df.columns.tolist())
        print("Primeiras linhas:")
        display(df.head())
else:
    print("❌ Nenhuma cidade foi identificada ou processada")
    print("Estrutura das colunas:", df.columns.tolist())
    print("Amostra dos dados:")
    display(df.head())
```
